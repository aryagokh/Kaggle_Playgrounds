```python
from autogluon.tabular import TabularDataset, TabularPredictor

train = TabularDataset('/kaggle/input/playground-series-s4e9/train.csv').drop('id', axis=1)
label = 'price'


predictor = TabularPredictor(label=label, 
                             problem_type='regression',
                             eval_metric='rmse',
                             verbosity=2).fit(
                                 train, 
                                 time_limit=3600*7.5,
                                 num_bag_folds=20,  # Increase bagging folds
                                 num_bag_sets=5,    # Increase bagging sets
                                 num_stack_levels=5,  # Increase stack levels
#                                  excluded_model_types=['KNN']  # If you want to keep KNN excluded
                             )
```

```python
print("Done")
```

```python
predictor.leaderboard()
```

```python
test = TabularDataset('/kaggle/input/playground-series-s4e9/test.csv').drop('id', axis=1)
pred1 = predictor.predict(test)
```

```python
sub = pd.read_csv("/kaggle/input/playground-series-s4e9/sample_submission.csv")
sub['price'] = pred1
sub.to_csv("1st-6-5-24.csv", index=False)
```
